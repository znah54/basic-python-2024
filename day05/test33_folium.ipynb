## folium 계속


```python
# file : test33_folium.ipynb
# desc : 포리움 지도 계속
import folium
```

```python
## 타일종류, 마커표시
#tiles 종류 => openstreetmap, cartodb positron, cartodb dark_mattr, open topomap
map = folium.Map(location=[35.173999, 129.196649], zoom_start=15, tiles='cartodb positron')
folium.Marker([35.173999, 129.196649], popup='송정해수욕장', ).add_to(map) # 마커추가
map
```

```python
## 마커 종류 변경
## [사랑해 다미~♡(성명건)] [오전 11:28] https://getbootstrap.com/docs/3.3/components/
map = folium.Map(location=[35.173999, 129.196649], zoom_start=15, tiles='cartodb positron')
folium.Marker([35.173999, 129.196649], popup='송정해수욕장',
               icon=folium.Icon(color='red', icon='remove-circle')).add_to(map) # 마커추가, 마커종류 변경
map
```
